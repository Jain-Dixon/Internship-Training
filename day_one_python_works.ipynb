```python
def count_vowels(s):
  return sum(1 for char in s.lower() if char in "aeiou")
result=count_vowels("jain")
print(result)

```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))   # Output: True
print(is_prime(10))  # Output: False
```

```python
def count_char(s, char):
    return s.count(char)

print(count_char("banana", "a"))   # Output: 3
print(count_char("hello", "l"))    # Output: 2
```

```python
def count_odd_digits(n):
    return sum(1 for d in str(abs(n)) if int(d) % 2 != 0)

print(count_odd_digits(1234567))   # Output: 4
print(count_odd_digits(2468))      # Output: 0
```

```python
def remove_spaces(s):
    return s.replace(" ", "")

print(remove_spaces("Data Science"))       # Output: DataScience
print(remove_spaces("Hello World Here"))
```

```python
def smallest(a, b, c):
    return min(a, b, c)

print(smallest(10, 5, 8))    # Output: 5
print(smallest(100, 3, 47))
```
